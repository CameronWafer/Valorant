```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stats_all = pd.read_excel("statsAllPlayerTeam.xlsx")
```

```python
#stats_all.drop("Unnamed: 0", axis=1, inplace=True)
stats_all.shape
```

```python
# get the top 25 rows with the highest K:D ratio
top_25_kd = stats_all.nlargest(25, 'K:D')

print(top_25_kd)
```

```python
top_25_players = top_25_kd["Player_Name"]
top_25_players
```

```python
names = ["azys", "florescent", "mmonch", "srN", "Moh", "iZu", "Korsonix", "Munchkin", "Verno", "SpYzera", 
         "Fair", "Noia", "consz", "Papi", "dods", "Meteor", "Alexy", "Vuubit", "madv", "ellie", 
         "Jade", "wippie", "sarahcat", "Persia", "karie"]

top_25 = stats_all[stats_all["Player_Name"].isin(names)]
top_25 = top_25[top_25["Player"] != "Papi  NBG"]
top_25
```

```python
percentage_columns = ['KAST', 'HS%', 'CL%']
for col in percentage_columns:
    top_25[col] = top_25[col].str.rstrip('%').astype(float) / 100
```

```python
# summary stats for top 25 players
top_25[['ACS', 'ADR', 'KAST', 'CL%']].describe()
```

```python
# correlation matrix
correlation_matrix = top_25[['K:D', 'ACS', 'ADR', 'KAST', 'CL%']].corr()

# visualize with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Key Metrics')
plt.show()
```

```python
# plotting distributions
top_25['ACS'].hist(bins=10)
plt.title('ACS Distribution for Top 25 Players')
plt.xlabel('ACS')
plt.ylabel('Frequency')
plt.show()
```

```python
# remove '%' from columns and convert to float
stats_all['KAST'] = stats_all['KAST'].str.rstrip('%').astype(float) / 100
stats_all['CL%'] = stats_all['CL%'].str.rstrip('%').astype(float) / 100
```

```python
# define the composite formula (IMPORTANT PART)
stats_all['Composite_Score'] = (
    stats_all['ACS'] * 0.4 +
    stats_all['ADR'] * 0.1 +
    stats_all['KAST'] * 0.3 +
    stats_all['CL%'] * 0.2
)

# rank players by composite score
top_composite = stats_all.nlargest(25, 'Composite_Score')

# display the top 25 players by composite score
print(top_composite[['Player_Name', 'Composite_Score']])
```

```python
# identify rising stars
rising_stars = stats_all[~stats_all['Player_Name'].isin(top_25['Player_Name'])].nlargest(25, 'Composite_Score')

# display rising stars
print(rising_stars[['Player_Name', 'Composite_Score']])
```
